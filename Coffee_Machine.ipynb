```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}
```

```python
def print_resources(resources):
    for key in resources.keys():
        print(f"{key}: {resources[key]}")

def user_choice():
    while True:
        user = input("What would you like?: ").lower()
        if user == "report":
            print_resources(resources)
        elif user in ["latte", "espresso", "cappuccino"]:
            return user
        else:
            print("Invalid choice. Please try again.")

def total_paid(quaters, dimes, nickles, pennies):
    quaters_cents = quaters * 0.25
    dimes_cents = dimes * 0.10
    nickles_cents = nickles * 0.05
    pennies_cents = pennies * 0.01
    sum_total = quaters_cents + dimes_cents + nickles_cents + pennies_cents
    return sum_total

def menu_cost(choice):
    cost = MENU[choice]["cost"]
    return cost

def can_make_drink(drink, menu, resources):
    ingredients = menu[drink]["ingredients"]

    for item, amount in ingredients.items():
        if item in resources:
            if resources[item] < amount:
                print(f"Sorry there is not enough {item}.")
                return False
        else:
            print(f"Sorry there is no {item}.")
            return False

    return True

def make_drink(drink, menu, resources):
    ingredients = menu[drink]["ingredients"]

    for item, amount in ingredients.items():
        resources[item] -= amount

def cashing(cost, total_paid):
    if total_paid < cost:
        print("Sorry that's not enough money. Money Refunded.")
        return False
    else:
        change = round(total_paid - cost, 2)
        print(f"Here is ${change} in change.")
        return True

while True:  # This loop will keep running until the user decides to stop
    choice = user_choice()

    if can_make_drink(choice, MENU, resources):
        print("Please insert coins.")
        quaters = float(input("How many quaters?: "))
        dimes = float(input("How many dimes?: "))
        nickles = float(input("How many nickles?: "))
        pennies = float(input("How many pennies?: "))

        total = total_paid(quaters, dimes, nickles, pennies)
        cost = menu_cost(choice)

        if cashing(cost, total):
            make_drink(choice, MENU, resources)
            print(f"Here is your {choice.capitalize()}. Enjoy!")
        else:
            print(f"Can't make {choice.capitalize()}.")
    else:
        print(f"Can't make {choice.capitalize()}.")

    another_drink = input("Do you want another drink? (yes/no): ").lower()
    if another_drink != "yes":
        break  # If the user doesn't want another drink, break the loop
```
